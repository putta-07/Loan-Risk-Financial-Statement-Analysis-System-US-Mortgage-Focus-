```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("/content/loan_financial_data.csv")
```

```python
# Data Cleaning
df.fillna(0, inplace=True)
```

```python
df.head()
```

```python
# Total Monthly Income
df["Total_Income"] = df["Applicant_Income"] + df["Coapplicant_Income"]

# Debt-to-Income Ratio (DTI)
df["DTI_Ratio"] = df["Existing_EMI"] / df["Total_Income"]

# Loan-to-Value Ratio (LTV)
df["LTV_Ratio"] = df["Loan_Amount"] / df["Property_Value"]
```

```python
# Risk Assessment Logic
def credit_risk(score):
    if score >= 750:
        return "Low Risk"
    elif score >= 650:
        return "Medium Risk"
    else:
        return "High Risk"

df["Credit_Risk"] = df["Credit_Score"].apply(credit_risk)
```

```python
# Loan Decision Logic

def loan_decision(row):
    if (
        row["Credit_Risk"] == "Low Risk"
        and row["DTI_Ratio"] <= 0.40
        and row["LTV_Ratio"] <= 0.80
    ):
        return "Approved"
    elif (
        row["Credit_Risk"] == "Medium Risk"
        and row["DTI_Ratio"] <= 0.45
        and row["LTV_Ratio"] <= 0.85
    ):
        return "Review Required"
    else:
        return "Rejected"

df["Loan_Decision"] = df.apply(loan_decision, axis=1)
```

```python
# Final Output

final_columns = [
    "Application_ID",
    "Total_Income",
    "DTI_Ratio",
    "LTV_Ratio",
    "Credit_Score",
    "Credit_Risk",
    "Loan_Decision"
]

final_report = df[final_columns]

print("\nFinal Loan Assessment Report:\n")
print(final_report)
```

```python
# Save final report
final_report.to_csv("loan_assessment_report.csv", index=False)
```
